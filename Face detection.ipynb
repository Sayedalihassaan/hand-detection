```python
import cv2
import mediapipe as mp
```

```python
mp_drawing_styles = mp.solutions.drawing_styles  

cap = cv2.VideoCapture(0)
mp_face_mesh = mp.solutions.face_mesh 

faces = mp_face_mesh.FaceMesh()

mp_drawing = mp.solutions.drawing_utils  


while True:
    # Read frame from camera
    ret, frame = cap.read()

    # Handle frame reading errors
    if not ret:
        print("Error reading frame")
        continue  
        
    # Convert image to RGB format (required by MediaPipe)
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame with the face mesh solution
    results = faces.process(image_rgb)
    print(results.multi_face_landmarks)

    # Draw face landmarks if detected
    if results.multi_face_landmarks is not None:
        for face_landmarks in results.multi_face_landmarks:
            mp_drawing.draw_landmarks(frame,  # Draw on the original frame
                                     face_landmarks,
                                     mp_face_mesh.FACEMESH_TESSELATION, 
                                     landmark_drawing_spec=None,
                                     connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_tesselation_style())  # Use appropriate style

    # Display the image with face landmarks
    cv2.imshow("MediaPipe Face Mesh", frame) 

    # Exit if 'q' key is pressed
    if cv2.waitKey(1) == ord("q"):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
